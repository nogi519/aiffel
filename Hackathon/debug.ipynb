```python
# 테스트용
import numpy as np
import os
from tensorflow import keras
from PIL import Image
import cv2
import tensorflow as tf
file_path = os.getenv('HOME') + '/beef/'
model_path = file_path + 'resnet50'
model = keras.models.load_model(model_path+'_aug.h5')


def predict(img,model=model):
    img = img.reshape(1,256,256,3)
    img = tf.cast(img, tf.float32)/255.
    return np.array(model(img)*100,dtype=np.float16), model(img).numpy().argmax()

img_path = os.getenv('HOME') + '/beef_data/All/test.jpg'
img = Image.open(img_path)
img = np.array(img.resize((256,256))).reshape(256,256,3)
# img2 = cv2.imread(img_path).reshape(256,256,3)
print(predict(img))
# predict(img2)
# np.array(img.resize((256,256))).reshape(256,256,3)
```
